```
df.filter(df['mobile'] == 'Vivo').show()
```

```
consulta1 = df.filter(df['mobile'] == 'Vivo')
consulta1.show() 
```

```
consulta2 = df.filter(df['mobile'] == 'Vivo').select('age', 'ratings')
consulta2.show() 
```

```
consulta3 = df.filter(df['mobile'] == 'Vivo').filter(df['ratings'] >= 4)
consulta3.show() 
```

```
consulta3 = df.filter((df['mobile'] == 'Vivo') | (df['ratings'] >= 4))
consulta3.show() 
# &, |, !=
```

```
# Ordenar por experiencia consulta anterior
#consulta4 = df.filter((df['mobile'] == 'Vivo') | (df['ratings'] >= 4)).orderBy('experience')
consulta4 = df.filter((df['mobile'] == 'Vivo') | (df['ratings'] >= 4)).sort('experience')
consulta4.show() 
# &, |, !=
```

```
consulta5 = df.filter((df['mobile'] == 'Vivo') | (df['ratings'] >= 4)).sort('experience', ascending=False)
consulta5.show()
```

```
# Ordenar por la experiencia y por la edad
consulta6 = df.filter((df['mobile'] == 'Vivo') | (df['ratings'] >= 4)).sort(df['experience'], df['age'])
consulta6.show()
```

```
# Ordenar por la experiencia y por la edad (descendente)
consulta7 = df.filter((df['mobile'] == 'Vivo') | (df['ratings'] >= 4)).sort(df['experience'].asc(), df['age'].desc())
consulta7.show()
```

```
# Calcular el promedio de la experiencia
df.groupBy('mobile').agg({'experience':'mean'}).show()
```

```
# Calcular el promedio de la experiencia
consulta8=df.groupBy('mobile').agg({'experience':'mean', 'ratings':'sum'})
consulta8.show()
```

```
consulta8 = consulta8.withColumnRenamed('avg(experience)','prom_experiencia').withColumnRenamed('sum(ratings','suma_ratings')
consulta8.show()
```
